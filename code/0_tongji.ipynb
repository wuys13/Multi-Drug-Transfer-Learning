```python
#每个癌症label数据数量（count）；和用药是否出现在CCL_dataset（top是出现多还是不出现多）
import pandas as pd
dataset = "gdsc1_raw"
Dataset_path = '../data/raw_dat/CCL_dataset/{}.csv'.format(dataset)
sensitivity_df = pd.read_csv(Dataset_path,index_col=1)
sensitivity_df.dropna(inplace=True,subset=['cid'])
ccle_sample_with_gex = pd.read_csv("../data/preprocessed_dat/ccle_sample_with_gex.csv",index_col=0)
target_df = sensitivity_df.loc[sensitivity_df.index.isin(ccle_sample_with_gex.index)]
ccl_cid_list = target_df[['Drug_name','cid']].drop_duplicates()
ccl_cid_list.index = range(1,ccl_cid_list.shape[0]+1)
ccl_cid_list.to_csv('../r_plot/data/labelled/ccl_cid_list.csv')

tcga = pd.read_csv('../data/tcga/PDR_data/tcga698_single_drug_response_df.csv',index_col=1)
xena_sample_with_gex = pd.read_csv("../data/preprocessed_dat/xena_sample_with_gex.csv",index_col=0)
xena_sample_with_gex.index = xena_sample_with_gex.index.map(lambda x: x[:12])
tcga = tcga.loc[tcga.index.isin(xena_sample_with_gex.index)]
# tcga
# xena_sample_with_gex

tcga['isin'] = tcga['cid'].isin(ccl_cid_list.cid.values)
tcga.to_csv('../r_plot/data/labelled/tcga698_PFS.csv')
tcga.groupby('tcga_project')['isin'].describe()  #  <5
```

```python
from torch.utils.data import TensorDataset, DataLoader
import torch
import numpy as np

a = np.array(range(10))
b = np.array([11,12,13,14])

dataset1 = TensorDataset(
        torch.from_numpy(a)
    )
dataset2 = TensorDataset(
        torch.from_numpy(b)
    )
dataloader1 = DataLoader(dataset1,
                                 batch_size=2,
                                 shuffle=True)
dataloader2 = DataLoader(dataset2,
                                 batch_size=2,
                                 shuffle=True)

for _,i in enumerate(dataloader1): 
    j = next(iter(dataloader2))
    print(i)
    print(j)
```
